# LIST

- list is ordered data type
- it stores multiple items in different data types
- its mutable: can be updated and added or removed data

## Key Features
- **ordered:** can be accessed by index. index starts from **0 to -1**
- **mutable:** elements can be changed, deleted and updated
- **Heterogeneous:** different types of data  can be stored
- **Dynamic Size:** grow and shrink any time
