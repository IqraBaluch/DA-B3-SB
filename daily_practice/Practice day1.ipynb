```python
#loop
import random

while True:
  choose_y_n = input("Enter (n/y) is you want to roll the dice: ").lower()

  if choose_y_n == 'y':
    die1 = random.randint(1,6)
    die2 = random.randint(1,6)
    print(f"{die1}, {die2}")

  elif choose_y_n == 'n':
    print("Thanks for playing.")
    break

  else:
    print("Invalid choice")

```

```python
# guess the number
# generate  random no
#input int
# if num is high [print too high]
# if num is low too low
#if num == num print match and win
# else in valid no
import random
rand = random.randint(1,10)
while True:
  try:
    guess = int(input("Enter a number btwn 1-10: "))

    if rand < guess:
      print("High")
    elif rand > guess:
      print("LOW")
    else:
      print("Congrats!!! you got it.")
      break

  except ValueError:
    print("Invalid input. Please enter a number.")

```
