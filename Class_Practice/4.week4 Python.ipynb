# IF- ELSE

````python
#single light example

light = 'green'

if light == 'red':
  print('stop') #action required
else:
  print('go')

````

````python

light = 'yellow'

if light == 'red':
  print('stop') #action required
elif light == 'yellow':
  print('slow down')
else:
  print('go')
````

````python
#grading

marks = 88
if marks >= 90:
  print('+A')
elif marks >= 80:
  print('A')
elif marks >= 70:
  print('B')
elif marks >=60:
  print('C')
elif marks >= 50:
  print('D')
else:
  print('fail')
````

````python
vowel = input("Enter a number to check if it is vowel or not: ")

if vowel.lower() == 'a' or vowel.lower()  == 'e' or vowel.lower()  == 'i' or vowel.lower()  == 'o' or vowel.lower()  == 'u':
  print("VOWEL")
else:
  print("CONSONANT")
````
